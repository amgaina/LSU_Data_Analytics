```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

parishes = ['Iberville Parish', 'Richland Parish', 'Catahoula Parish', 'De Soto Parish', 'Avoyelles Parish']
cancer_rates = [780.5, 759.5, 756.6, 743.5, 741.5]

df = pd.DataFrame({
    'Parish': parishes,
    'Cancer Rate': cancer_rates
})
palette = sns.color_palette("YlOrRd", len(df))
plt.figure(figsize=(10, 6))
sns.barplot(x='Parish', y='Cancer Rate', data=df, palette=palette)

plt.title('Top 5 Parishes by Cancer Rate for black males(per 100,000)', fontsize=16, weight='bold', pad=20)
plt.xlabel('Parish', fontsize=14)
plt.ylabel('Cancer Rate', fontsize=14)

for index, value in enumerate(df['Cancer Rate']):
    plt.text(index, value + 5, f'{value:.1f}', ha='center', fontsize=12, weight='bold')

plt.xticks(rotation=15, fontsize=12)  # Rotating parish names for better readability
plt.yticks(fontsize=12)
plt.grid(True, which='major', axis='y', linestyle='--', linewidth=0.7, alpha=0.7)

sns.despine()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
df_female_state = df1.sort_values(by = "All Sites", ascending = False).head(5)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
parishes_female = df_female_state["State"]
cancer_rate_female = df_female_state["All Sites"]

df_female = pd.DataFrame({
    'Parish': parishes_female,
    'Cancer Rate (Female)': cancer_rate_female
})

palette = sns.color_palette("Purples", len(df_female))


plt.figure(figsize=(10, 6))
sns.barplot(x='Parish', y='Cancer Rate (Female)', data=df_female, palette=palette)


plt.title('Cancer Rates by Parish (Females)', fontsize=16, weight='bold', pad=20)
plt.xlabel('Parish', fontsize=14)
plt.ylabel('Cancer Rate (per 100,000)', fontsize=14)

# Adding values on the bars
for index, value in enumerate(df_female['Cancer Rate (Female)']):
    plt.text(index, value + 5, f'{value:.1f}', ha='center', fontsize=12, weight='bold')

plt.xticks(rotation=15, fontsize=12)  # Rotating parish names for better readability
plt.yticks(fontsize=12)
plt.grid(True, which='major', axis='y', linestyle='--', linewidth=0.7, alpha=0.7)

sns.despine()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
df3 = pd.read_excel("percent_cancer.xlsx", sheet_name = "black_percent")
```

```python
type(df3["Male"])
```

```python
df3["Male"] = df3["Male"].astype(float)
```

```python
df3.drop(index = 0, inplace = True)
```

```python
df3.sort_values(by = "Male", ascending = False)
```

```python
df3.sort_values(by = "Male", ascending = False).head()
```

```python
cancer_types_male = ['Male Genital System', 'Prostate', 'Digestive System', 'Respiratory System', 'Lung and Bronchus']
percentages_male = [34.9, 34.3, 23.0, 16.6, 14.6]

colors = sns.color_palette('Set3', len(cancer_types_male))

plt.figure(figsize=(8, 8))
wedges, _ = plt.pie(percentages_male, labels=None, colors=colors, 
                     startangle=90, wedgeprops=dict(width=0.3))

centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Top 5 Cancer Types Percentage Distribution in Black Male Population (Louisiana)', fontsize=16, weight='bold')

plt.legend(wedges, [f'{cancer_types_male[i]}: {percentages_male[i]}%' for i in range(len(cancer_types_male))],
           title="Cancer Types & Rates", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1), fontsize=12)

plt.tight_layout()
plt.show()
```
